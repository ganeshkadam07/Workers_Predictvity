# df2 is a copy dataset of df

```python
df2=df.copy()

df2.head()
```

# Drop wip column

```python
df2.drop(['wip'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
# Split dataset into Training and Testing set

x=df2.iloc[:,2:6].values
y=df2.iloc[:,-1].values

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)

# shape of Training and Testing set
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
# Scale The values using Standardscalar

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
```

```python
# Fit Training and Testing data into Standardscalar

x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

# With HyperParameters

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [100, 150, 200, 250, 300],
    'max_depth': [1,2,3,4],
    'min_samples_split': [1,2,3,4],
    'max_features': [1.0,2.0,3.0]
}
regr = RandomForestRegressor(random_state=0)

clf = GridSearchCV(regr, parameters)
clf.fit(x_train, y_train)
```

**Finding Best Parameter**

```python
clf.best_params_
```

# Making a Model Using Parameters

```python
rf_model=RandomForestRegressor(max_depth= 4,
 max_features= 1.0,
 min_samples_split= 2,
 n_estimators= 200)

rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_test)
```

# **Calculate Mean Squared Error**

```python
# Find Mean_squared_Error 

from sklearn.metrics import mean_squared_error

print("Mean Squared Error:",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor


parameters={
            'learning_rate': [0.0,1.0,2.0,3.0],
            'n_estimators': [100,150,200,250,300],
            'min_samples_leaf': [1,2,3,4],
            'max_depth': [1,2,3,4,5],
            'max_features': [1.0,2.0,3.0],
            }


GB_model=GradientBoostingRegressor(random_state=0)

GB_model= GridSearchCV(GB_model, parameters)
GB_model.fit(x_train, y_train)
```

```python
GB_model.best_params_
```

```python
Grad_model=GradientBoostingRegressor(learning_rate= 1.0,
 max_depth= 2,
 max_features= 1.0,
 min_samples_leaf= 4,
 n_estimators= 300)

Grad_model.fit(x_train,y_train)
```

```python
gred_pred=Grad_model.predict(x_test)
```

# **Mean_Squared_Error of GradientBoosting Model**

```python
# Calculate mean_squared_error of Gradient Boosting Regressor model

print("mean Squared Error:",mean_squared_error(y_test,gred_pred))
```

# 3] XGBoost Regressor

```python
from xgboost import XGBRegressor

xg_param={'n_estimators':[100,150,200,250,300],
          'max_depth': [1,2,3,4,5],
          'gamma': [1,2,3,4,5],
          'reg_alpha': [20,40,60,80,100,120],
          'min_child_weight': [1,2,3,4,5],
          }

X_model=XGBRegressor()

xg_clf = GridSearchCV(X_model, xg_param)
xg_clf.fit(x_train, y_train)
```

```python
xg_clf.best_params_
```

```python
x_model=XGBRegressor(gamma= 1,
 max_depth= 1,
 min_child_weight= 1,
 n_estimators= 100,
 reg_alpha= 20)

x_model.fit(x_train,y_train)
```

```python
X_pred=x_model.predict(x_test)
```

# Mean Squared Error of XGBoost Regressor

```python
# mean Squared error

print("MSE:",mean_squared_error(y_test,X_pred))
```
